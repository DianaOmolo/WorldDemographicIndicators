## 1. World

```python
#Creating a dataframe for group "World"
world_df = df[df.Type =='World']

world_df
```

```python
world_df.describe()
```

```python
#What year had the highest population growth rate in the world?
world_df.sort_values('Population Growth Rate', ascending=False).head(3)

#1963
```

```python
#What year had the lowest population growth rate?
world_df.sort_values('Population Growth Rate').head(3)

#2021
```

### *The highest population growth rate  in the world between 1950 and 2021 was in 1963 at 2.27% while the lowest was in 2021 at 0.82%. From this we can assume that over the years the population growth rate has really decreased.*

```python
# What year had the largest number of births in the world?
world_df.sort_values('Births', ascending=False).head(3)

#2012 had the highest number of births at 144,194,000
```

```python
#checking the correlation between Births and Total Population

plt.title('Correlation Between Births and The Total Population');
sns.regplot(x='Births', y='Total Population', data= world_df);

plt.figure(figsize=(10,6));


# They are highly correlated
```

```python
# Visualization on some of the demographic indicators to give an overall view of the changes that occurred (1950-2021)

fig, axes = plt.subplots(2, 3, figsize=(16, 8))

axes[0,0].plot(world_df['Year'], world_df['Total Population'])
axes[0,0].set_title('Total Population');

axes[0,1].plot(world_df['Year'], world_df['Total Deaths'])
axes[0,1].set_title('Total Deaths');

axes[0,2].plot(world_df['Year'], world_df['Fertility Rate'])
axes[0,2].set_title('Fertility Rate');

axes[1,0].plot(world_df['Year'], world_df['Life Expectancy'])
axes[1,0].set_title('Life Expectancy');

axes[1,1].plot(world_df['Year'], world_df['Infant Mortality Rate'])
axes[1,1].set_title('Infant Mortality Rate');

axes[1,2].plot(world_df['Year'], world_df['Under 5 Mortality Rate'])
axes[1,2].set_title('Under 5 Mortality Rate');

plt.tight_layout(pad=2);
```

### *1. The world's population has increased steadily between 1950 and 2021.*
### *2. The number of deaths drastically increased in 2020 definitely due to the COVID-19 Pandemic.*
### *3. The fertility rate in women has really decreased in the 71 years.*
### *4. Life expectancy steadily increased between 1960 and 2019 and then a drastic drop indicating decrease.*
### *5. Infant and Under 5 mortality rates have majorly decreased in the 71 years which is an amazing thing.*

## 2. SDG Region(Sustainable Development Goals)
### *Under this group i'll focus on the year 2021*

```python
sdg_df = df[df.Type =='SDG region']

sdg_df.sample(5)
```

```python
sdg_df.describe()
```

```python
sns.boxplot(x='Year', y='Births by womenaged 15 to 19', data= sdg_df)

sns.set(rc={'figure.figsize': (16,6)})

plt.title('Births by Women Aged 15 to 19');
plt.xticks(rotation=90);
```

```python
# The distinct regions the countries have been grouped into according to the Sustainable Development Goals
sdg_df['Region, subregion, country or area '].unique()
```

```python
year_df=sdg_df[sdg_df.Year==2021] 

year_df.sort_values('Total Population', ascending=False)
```

### *In the year 2021 Eastern and South-Eastern Asia SDG Region had the highest population and Oceania(excluding Australia and New Zealand) SDG Region had the lowest population. This is dependent on how many countries make up a region and the population of individual countries.*

```python
pop= year_df.groupby('Region, subregion, country or area ')[['Fertility Rate',
                                                            'Crude Death Rate', 'Life Expectancy']].sum()
pop.sort_values(['Fertility Rate', 'Crude Death Rate'])
```

### *In 2021, Sub-Saharan Africa SDG Region had the highest Fertility Rate, Europe and Northern America SDG Region had the highest death rate, and  Australia/New Zealand SDG Region had the highest Life expectancy.*

```python
# Finding out the correlation between a few indicators.
correlation_matrix= sdg_df[['Total Population', 'Births', 'Population Growth Rate','Fertility Rate']].corr()

correlation_matrix
```

### *Total Population is highly correlated to Births and Population Growth Rate is highly correlated to Fertility Rate.*

```python
plt.figure(figsize=(8,4));
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn');

plt.title('Correlation Matrix for Numeric Values');

plt.show();
```

## 3. Income Group

```python
income_df = df[df.Type =='Income Group']

income_df.sample(5)
```

```python
income_df.describe()
```

```python
income_df['Region, subregion, country or area '].unique()
```

```python
# Looking at the maximum of a few indicators of income groups irrespective of the year
inc_df= pd.pivot_table(income_df, values=['Total Population', 'Population Growth Rate',
                                          'Life Expectancy', 'Crude Death Rate', 'Infant Mortality Rate'], 
                    index='Region, subregion, country or area ',aggfunc=max).sort_values('Crude Death Rate', ascending= False)

inc_df
```

### Looking at High-income countries and Low-income countries maximum values between 1950 and 2021 we can see that in Crude Death Rate and Infant Mortality Rate, Low-income countries were leading. In Life expectancy High-income countries had the highest.
